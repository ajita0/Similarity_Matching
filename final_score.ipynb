```python
# Removing na values from dataframe
def data_na_value_cleaning(data):
    print("\nBefore cleaning, Data Shape : ", data.shape)
    print("\nBefore removing Null values: ---------------")
    print(data.isna().sum())
    
    data.dropna(inplace=True)
    data.reset_index(inplace=True,drop=True)
    
    print("After removing Null values: ---------------")
    print(data.isna().sum())
    print("\nAfter cleaning, Data Shape : ", data.shape)
    
    return data
```

```python
# Removing duplicate values
def duplicate_content_removal(data, col, ini_row):
    print("\nBefore removing duplicates, number of data was : ", ini_row)
    duplicate_count = data[col].duplicated().sum()
    print("\nNumber of Duplicates: ", duplicate_count)

    description_data = data[col].drop_duplicates()
    cleaned_row = len(description_data)
    
    if (ini_row - cleaned_row) > 0:
        print("\nTotal data reduction : ", (ini_row - cleaned_row))
        print("\nAfter removing duplicates, number of data is :", cleaned_row)
    else:
        print("\nDataset doesn't content any duplicate data.")
    
    return list(description_data)
```

```python
def load_resumes(folder_path):
    resumes = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            with open(os.path.join(folder_path, filename), "r", encoding="utf-8") as file:
                resume = file.read()
                resumes.append(resume)
    return resumes

def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    word_tokens = nltk.word_tokenize(text.lower())
    filtered_words = [word for word in word_tokens if word not in stop_words]
    return filtered_words
```

```python
import re
import os

def extract_education_section(file_path):
    # Read the text file
    with open(file_path, 'r') as file:
        resume_text = file.read()

    # Define keywords to identify the education section
    education_keywords = ["EDUCATION", "EDUCATIONAL BACKGROUND"]

    # Initialize variables
    education_section = ""

    # Search for education section using keywords
    for keyword in education_keywords:
        keyword = keyword.upper()
        match = re.search(r"\b" + re.escape(keyword) + r"\b", resume_text.upper())
        if match:
            start_index = match.end()
            end_index = resume_text.find("\n\n", start_index)
            education_section = resume_text[start_index:end_index].strip()
            break

    return education_section

# Directory containing the resume files
resumes_directory = r'C:\Users\user\Desktop\Domain_IT\Resume'

# Extract education section from each resume
for filename in os.listdir(resumes_directory):
    if filename.endswith(".txt"):
        resume_file_path = os.path.join(resumes_directory, filename)
        education_section = extract_education_section(resume_file_path)
        print(f"Resume: {filename}")
        print(f"Education Section: {education_section}")
        print("-------------------------------------")
```

```python
import re
import os

def extract_qualification_training_section(file_path):
    # Read the text file
    with open(file_path, 'r') as file:
        resume_text = file.read()

    # Define keywords to identify the qualification and training sections
    qualification_keywords = ["QUALIFICATIONS", "ACADEMIC QUALIFICATIONS"]
    training_keywords = ["TRAINING", "CERTIFICATIONS"]

    # Initialize variables
    qualification_section = ""
    training_section = ""

    # Search for qualification and training sections using keywords
    for keyword in qualification_keywords:
        keyword = keyword.upper()
        match = re.search(r"\b" + re.escape(keyword) + r"\b", resume_text.upper())
        if match:
            start_index = match.end()
            end_index = resume_text.find("\n\n", start_index)
            qualification_section = resume_text[start_index:end_index].strip()
            break

    for keyword in training_keywords:
        keyword = keyword.upper()
        match = re.search(r"\b" + re.escape(keyword) + r"\b", resume_text.upper())
        if match:
            start_index = match.end()
            end_index = resume_text.find("\n\n", start_index)
            training_section = resume_text[start_index:end_index].strip()
            break

    return qualification_section, training_section

# Directory containing the resume files
resumes_directory = r'C:\Users\user\Desktop\Domain_IT\Resume'

# Extract qualification and training sections from each resume
for filename in os.listdir(resumes_directory):
    if filename.endswith(".txt"):
        resume_file_path = os.path.join(resumes_directory, filename)
        qualification_section, training_section = extract_qualification_training_section(resume_file_path)
        print(f"Resume: {filename}")
        print(f"Qualification Section: {qualification_section}")
        print(f"Training Section: {training_section}")
        print("-------------------------------------")
```

```python
import numpy as np
#similarity = np.array([])
def calculate_similarity(resume, job_description, experience_weight=0.3, skills_weight=0.4, education_weight=0.1,
                         qualification_training_weight=0.2):
    # Preprocess text
    resume_tokens = preprocess_text(resume)
    job_desc_tokens = preprocess_text(job_description)

    # Load Word2Vec model
    model = Word2Vec(job_desc_tokens + resume_tokens, sg=0, window=5, min_count=1, workers=4, seed=42)
    model.train(job_desc_tokens + resume_tokens, total_examples=model.corpus_count, epochs=5)

    # Calculate weighted average vector for resume
    resume_vector = (experience_weight * sum(model.wv[word] for word in resume_tokens if word in model.wv)) / len(resume_tokens)

    # Calculate weighted average vector for job description
    job_vector = ((experience_weight * sum(model.wv[word] for word in job_desc_tokens if word in model.wv)) +
                  (skills_weight * sum(model.wv[word] for word in job_desc_tokens if word in model.wv)) +
                  (education_weight * sum(model.wv[word] for word in job_desc_tokens if word in model.wv)) +
                  (qualification_training_weight * sum(model.wv[word] for word in job_desc_tokens if word in model.wv))) / len(job_desc_tokens)

    # Apply experience, skills, education, qualification, and training weight to the resume vector
    resume_vector = (1 - experience_weight) * resume_vector

    # Calculate cosine similarity between resume and job description vectors
    similarity = cosine_similarity([resume_vector], [job_vector])[0][0]
    return similarity

# Specify the folder path containing job descriptions
job_descriptions_folder = r'C:\Users\user\Desktop\Domain_IT\Job_description'
#job_descriptions_folder = r'C:\Users\user\Desktop\Domain_Digital marketing\Job_Description'

# Specify the folder path containing resumes
resume_folder = r'C:\Users\user\Desktop\Domain_IT\Resume'

# Initialize empty lists
job_descs = []
resumes = []

# Read job descriptions from folder
for filename in os.listdir(job_descriptions_folder):
    with open(os.path.join(job_descriptions_folder, filename), "r", encoding="utf-8") as file:
        job_desc = file.read()
        job_descs.append(job_desc)

# Read resumes from folder
for filename in os.listdir(resume_folder):
    with open(os.path.join(resume_folder, filename), "r", encoding="utf-8", errors="ignore") as file:
        resume = file.read()
        resumes.append(resume)

# Load resumes
resumes = load_resumes(resume_folder)

# Choose the job description you want to compare with
job_description = job_descs[2]

# Iterate over resumes and calculate similarity with the job description
for resume in resumes:
    similarity = calculate_similarity(resume, job_description, experience_weight=0.3, skills_weight=0.4,
                                     education_weight=0.1, qualification_training_weight=0.2)
    #similarity = np.append(calculate_similarity(resume, job_description, experience_weight=0.3, skills_weight=0.4,
                                       #education_weight=0.1, qualification_training_weight=0.2))
    #similarity = np.sort(similarity)
    print("Similarity:", similarity)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

for resume in resumes:
    similarity = calculate_similarity(resume, job_description, experience_weight=0.3, skills_weight=0.4,
                                      education_weight=0.1, qualification_training_weight=0.2)

    # Determine if the resume is relevant or not based on the threshold (e.g., 0.5)
    if similarity >= 0.5:
        predicted_labels.append(1)  # Relevant
    else:
        predicted_labels.append(0)  # Irrelevant

    # Check if the resume is actually relevant (You need to have the ground truth labels for this)
    # If relevant, append 1; if irrelevant, append 0
    true_labels.append(1)  # Assuming the job description is always relevant

    # If you have ground truth labels for each resume, use them instead of the assumed value.

# Calculate true positives (TP), true negatives (TN), false positives (FP), and false negatives (FN)
conf_matrix = confusion_matrix(true_labels, predicted_labels, labels=[1, 0])

# Calculate classification report for precision, recall, F1-score, and support metrics
report = classification_report(true_labels, predicted_labels, labels=[1, 0], target_names=['Relevant', 'Irrelevant'])

print("Confusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(report)
```
